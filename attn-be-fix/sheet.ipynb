```python
import os

from dotenv import load_dotenv
import pymysql
# import pandas

load_dotenv()

DB_HOST = os.getenv("DB_HOST")
DB_USER = os.getenv("DB_USER")
DB_PASSWORD = os.getenv("DB_PASSWORD")
DB_SCHEMA = os.getenv("DB_SCHEMA")

con = pymysql.connect(host=DB_HOST, user=DB_USER, password=DB_PASSWORD, port=3306, database=DB_SCHEMA)

with pymysql.cursors.DictCursor(con) as cursor:
    cursor.execute("SELECT * FROM employees")
    employees = cursor.fetchall()

    for employee in employees:
        # we need empId, employeeName, and employeeDeploySite
        month = 4

        with pymysql.cursors.DictCursor(con) as attCursor:
            attCursor.execute("SELECT * FROM attendance_record WHERE emp_id = %s AND date BETWEEN %s AND %s", (employee['empId'], f"2024-{month if month > 9 else f'0{month}'}-01", f"2024-{month if month > 9 else f'0{month}'}-31"))

            attendances = attCursor.fetchall()

            for attn in attendances:
                print(attn)

con.close()
```
